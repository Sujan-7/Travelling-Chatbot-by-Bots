```python
class TravelChatbot:
    def __init__(self):
        self.data = {
            "madurai to goa": {
                "bus_timings": ["8:00 AM", "11:00 AM", "2:00 PM"],
                "hotel_timings": "24/7",
                "restaurant_timings": "7:00 AM - 11:00 PM",
                "lodges": ["Lodge 1", "Lodge 2"],
                "description": "A scenic route connecting madurai and goa."
            },
            # Add more place pairs and their data here
        }

    def handle_query(self, query):
        query = query.lower()
        for keyword, info in self.data.items():
            if keyword in query:
                return self.get_info(info)
        return "I'm sorry, I don't have information on that route."

    def get_info(self, info):
        bus_timings = ", ".join(info["bus_timings"])
        hotel_timings = info["hotel_timings"]
        restaurant_timings = info["restaurant_timings"]
        lodges = ", ".join(info.get("lodges", []))
        description = info.get("description", "No description available.")

        return (
            f"Description: {description}\n"
            f"Bus timings: {bus_timings}\n"
            f"Hotel timings: {hotel_timings}\n"
            f"Restaurant timings: {restaurant_timings}\n"
            f"Lodges available: {lodges}"
        )

    def start(self):
        print("Hello! I'm your travel chatbot.")
        print("You can ask me about buses, hotels, restaurants, and lodges between places.")
        print("For example: 'Tell me about madurai to goa'")
        user_input = input("You: ")
        response = self.handle_query(user_input)
        print("Chatbot:", response)
        print("Goodbye!")


if __name__ == "__main__":
    chatbot = TravelChatbot()
    chatbot.start()
```
